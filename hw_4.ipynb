```python
#Азбука

number = int(input())

def check_first_letter(words):
    allowed_letters = ['A', 'B', 'C']
    
    for word in words:
        first_letter = word[0].upper()  # Convert the first letter to uppercase
        
        if first_letter not in allowed_letters:
            return 'NO'
    
    return 'YES'
```

```python
#Анонс новости

def shorten_heading(heading, length):
    if len(heading) <= length:
        return heading  # No need to shorten
        
    shortened_heading = heading[:length-3] + "..."  # Subtracting 3 to account for the ellipsis
    return shortened_heading

# Input
L = int(input("Enter the required header length: "))
N = int(input("Enter the number of headings: "))

headings = []
for i in range(N):
    heading = input("Enter heading {}: ".format(i+1))
    headings.append(heading)

# Output
print("Abbreviated headings:")
for heading in headings:
    shortened = shorten_heading(heading, L)
    print(shortened)
```

```python
#Очистка данных

def clean_report(report_lines):
    cleaned_lines = []
    
    for line in report_lines:
        if line.startswith("##"):
            continue
        
        if line.endswith("@@@"):
            continue
        
        cleaned_lines.append(line)
    
    return cleaned_lines

# Input
report = []
while True:
    line = input()
    if line == "":
        break
    report.append(line)

# Output
cleaned_report = clean_report(report)
for line in cleaned_report:
    print(line)
```

```python
#Без комментариев

def remove_comments(program_lines):
    cleaned_lines = []
    
    for line in program_lines:
        line = line.strip()  # Remove leading/trailing whitespace
        
        if line.startswith("#"):  # Entire line is a comment
            continue
        
        if "#" in line:  # Comment is present in the line
            line = line[:line.index("#")]  # Remove the comment part
        
        cleaned_lines.append(line)
    
    return cleaned_lines

# Input
program = []
while True:
    line = input()
    if line == "":
        break
    program.append(line)

# Output
cleaned_program = remove_comments(program)
for line in cleaned_program:
    print(line)
```

```python
#Частотный анализ на минималках


def find_most_frequent_character(text):
    character_count = {}
    
    for char in text:
        if char.islower():  # Exclude uppercase, digits, and special characters
            character_count[char] = character_count.get(char, 0) + 1
    
    max_count = 0
    most_frequent_chars = []
    for char, count in character_count.items():
        if count > max_count:
            max_count = count
            most_frequent_chars = [char]
        elif count == max_count:
            most_frequent_chars.append(char)
    
    return sorted(most_frequent_chars)[0]

# Input
lines = []
while True:
    line = input()
    if line == "FINISH":
        break
    lines.append(line)

text = "".join(lines)

# Output
most_frequent_char = find_most_frequent_character(text)
print(most_frequent_char)
```

```python
#Массовое возведение в степень


def raise_to_power(numbers, power):
    powered_numbers = []
    
    for num in numbers:
        powered_num = num ** power
        powered_numbers.append(powered_num)
    
    return powered_numbers

# Input
N = int(input("Enter the number of numbers: "))

numbers = []
for i in range(N):
    num = float(input("Enter number {}: ".format(i+1)))
    numbers.append(num)

power = int(input("Enter the power: "))

# Output
powered_sequence = raise_to_power(numbers, power)
for num in powered_sequence:
    print(num)
```

```python
#НОД 3.0


def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Input
numbers = list(map(int, input().split()))

# Find GCD
result = numbers[0]
for i in range(1, len(numbers)):
    result = gcd(result, numbers[i])

# Output
print(result)
```

```python
#RLE

def rle_encode(digits):
    encoded_sequence = []
    count = 1
    
    for i in range(1, len(digits)):
        if digits[i] == digits[i - 1]:
            count += 1
        else:
            encoded_sequence.append((digits[i - 1], count))
            count = 1
    
    encoded_sequence.append((digits[-1], count))
    
    return encoded_sequence

# Input
digits = input("Enter a string of digits: ")

# RLE Encoding
encoded_pairs = rle_encode(digits)

# Output
for pair in encoded_pairs:
    print(pair[0], pair[1])
```

```python
#Символическая разница

def find_common_symbols(string1, string2):
    common_symbols = set(string1) & set(string2)
    return common_symbols

# Input
string1 = input("Enter the first string: ")
string2 = input("Enter the second string: ")

# Find Common Symbols
common_symbols = find_common_symbols(string1, string2)

# Output
for symbol in common_symbols:
    print(symbol)
```

```python
#Кашееды

def find_common_children(semolina_lovers, oatmeal_lovers):
    common_lovers = set(semolina_lovers) & set(oatmeal_lovers)
    return len(common_lovers)

# Input
N = int(input("Enter the number of children who love semolina: "))
M = int(input("Enter the number of children who love oatmeal: "))

semolina_lovers = []
oatmeal_lovers = []

for i in range(N):
    semolina_lover = input("Enter the name of child who loves semolina: ")
    semolina_lovers.append(semolina_lover)

for i in range(M):
    oatmeal_lover = input("Enter the name of child who loves oatmeal: ")
    oatmeal_lovers.append(oatmeal_lover)

# Find Common Children
common_children_count = find_common_children(semolina_lovers, oatmeal_lovers)

# Output
if common_children_count > 0:
    print(common_children_count)
else:
    print("There are no such people.")
```

```python
#Зайка

def count_objects(description):
    object_counts = {}

    for line in description:
        if line.strip() == "":
            break
        
        objects = line.split()
        for obj in objects:
            if obj in object_counts:
                object_counts[obj] += 1
            else:
                object_counts[obj] = 1

    return object_counts

# Input
description = []
while True:
    line = input()
    if line.strip() == "":
        break
    description.append(line)

# Count Objects
object_counts = count_objects(description)

# Output
for obj, count in object_counts.items():
    print(obj, count)
```

```python
#Транслитерация

def transliterate_text(text):
    replacements = {
        'А': 'A',
        'Б': 'B',
        'В': 'V',
        'Г': 'G',
        'Д': 'D',
        'Е': 'E',
        'Ё': 'E',
        'Ж': 'ZH',
        'З': 'Z',
        'И': 'I',
        'Й': 'I',
        'К': 'K',
        'Л': 'L',
        'М': 'M',
        'Н': 'N',
        'О': 'O',
        'П': 'P',
        'Р': 'R',
        'С': 'S',
        'Т': 'T',
        'У': 'U',
        'Ф': 'F',
        'Х': 'KH',
        'Ц': 'TS',
        'Ч': 'CH',
        'Ш': 'SH',
        'Щ': 'SHCH',
        'Ы': 'Y',
        'Э': 'E',
        'Ю': 'YU',
        'Я': 'YA',
        'а': 'a',
        'б': 'b',
        'в': 'v',
        'г': 'g',
        'д': 'd',
        'е': 'e',
        'ё': 'e',
        'ж': 'zh',
        'з': 'z',
        'и': 'i',
        'й': 'i',
        'к': 'k',
        'л': 'l',
        'м': 'm',
        'н': 'n',
        'о': 'o',
        'п': 'p',
        'р': 'r',
        'с': 's',
        'т': 't',
        'у': 'u',
        'ф': 'f',
        'х': 'kh',
        'ц': 'ts',
        'ч': 'ch',
        'ш': 'sh',
        'щ': 'shch',
        'ы': 'y',
        'э': 'e',
        'ю': 'yu',
        'я': 'ya'
    }

    transliterated_text = ""
    i = 0

    while i < len(text):
        if text[i] in replacements:
            transliterated_text += replacements[text[i]]
            if text[i].isupper():
                transliterated_text = transliterated_text.capitalize()
        elif text[i] in ['ъ', 'Ъ', 'ь', 'Ь']:
            i += 1
            continue
        else:
            transliterated_text += text[i]
        i += 1

    return transliterated_text

# Input
text = input()

# Transliterate Text
transliterated_text = transliterate_text(text)

# Output
print(transliterated_text)
```

```python
#Однофамильцы — 2

def find_namesakes(employees):
    namesakes = {}
    for employee in employees:
        if employee in namesakes:
            namesakes[employee] += 1
        else:
            namesakes[employee] = 1
    return namesakes

# Input
N = int(input("Enter the number of male employees: "))
employees = []
for _ in range(N):
    employee = input("Enter the name of an employee: ")
    employees.append(employee)

# Find Namesakes
namesakes = find_namesakes(employees)

# Output
if len(namesakes) > 0:
    for employee, count in sorted(namesakes.items()):
        print(employee, count)
else:
    print("No namesakes")
```

```python
#Простая задача 4.0

import math

def find_coprimes(numbers):
    coprimes = {}

    for number in numbers:
        coprimes[number] = []

        for i in range(2, number):
            if math.gcd(i, number) == 1:
                coprimes[number].append(i)

    return coprimes

# Input
numbers = input("Enter the sequence of numbers: ").split("; ")

# Find Coprimes
coprimes = find_coprimes(numbers)

# Output
for number in numbers:
    if len(coprimes[number]) > 0:
        coprime_list = ", ".join(str(coprime) for coprime in coprimes[number])
        print(f"{number} - {coprime_list}")
```

```python
#Друзья друзей

from collections import defaultdict

def find_second_level_friends(pairs):
    friendships = defaultdict(set)
    second_level_friends = {}

    for pair in pairs:
        person1, person2 = pair.split()
        friendships[person1].add(person2)
        friendships[person2].add(person1)

    for person, friends in friendships.items():
        second_level_friends[person] = set()
        for friend in friends:
            second_level_friends[person].update(friendships[friend])

        # Remove person and their direct friends from the second level friends
        second_level_friends[person].discard(person)
        second_level_friends[person].difference_update(friends)

    return second_level_friends

# Input
pairs = []
while True:
    pair = input()
    if pair == "":
        break
    pairs.append(pair)

# Find 2nd Level Friends
second_level_friends = find_second_level_friends(pairs)

# Output
for person, friends in sorted(second_level_friends.items()):
    friend_list = ", ".join(sorted(friends))
    print(f"{person}: {friend_list}")
```
