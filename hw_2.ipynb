```python
#Корень зла

import math

def find_roots(a, b, c):
    discriminant = b**2 - 4*a*c

    if discriminant > 0:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return [root1, root2]
    elif discriminant == 0:
        root = -b / (2*a)
        return [root]
    else:
        return []

# Input
a = float(input("Enter the coefficient a: "))
b = float(input("Enter the coefficient b: "))
c = float(input("Enter the coefficient c: "))

# Find Roots
roots = find_roots(a, b, c)

# Output
if len(roots) == 0:
    print("No solution")
elif len(roots) == 1:
    print(f"Root: {roots[0]:.2f}")
else:
    print(f"Roots: {roots[0]:.2f}, {roots[1]:.2f}")
```

```python
#Внимание! Акция!

def calculate_total_cost(prices):
    total_cost = sum(prices)

    if total_cost >= 500:
        discount = total_cost * 0.1
        total_cost -= discount

    return total_cost

# Input
prices = []
while True:
    price = float(input("Enter the price of a product (enter 0 to finish): "))
    if price == 0:
        break
    prices.append(price)

# Calculate Total Cost
total_cost = calculate_total_cost(prices)

# Output
print(f"Total purchase amount: {total_cost}")
```

```python
#НОК


def calculate_lcm(a, b):
    # Find the greater number
    if a > b:
        greater = a
    else:
        greater = b

    while True:
        if greater % a == 0 and greater % b == 0:
            lcm = greater
            break
        greater += 1

    return lcm

# Input
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))

# Calculate LCM
lcm = calculate_lcm(a, b)
```

```python
#Факториал

def calculate_factorial(n):
    factorial = 1

    if n == 0:
        return factorial

    for i in range(1, n + 1):
        factorial *= i

    return factorial

# Input
n = int(input("Enter a non-negative number: "))

# Calculate Factorial
factorial = calculate_factorial(n)

# Output
print(f"Factorial: {factorial}")
```

```python
#Цифровая сумма

def calculate_digit_sum(number):
    digit_sum = 0
    for digit in str(number):
        digit_sum += int(digit)
    return digit_sum

# Input
number = int(input("Enter a natural number: "))

# Calculate Digit Sum
digit_sum = calculate_digit_sum(number)

# Output
print(f"Sum of digits: {digit_sum}")
```

```python
#Простая задача


import math

def is_prime(number):
    if number <= 1:
        return False

    # Check divisibility up to the square root of the number
    for i in range(2, int(math.sqrt(number)) + 1):
        if number % i == 0:
            return False

    return True

# Input
number = int(input("Enter a natural number: "))

# Check if Prime
if is_prime(number):
    print("YES")
else:
    print("NO")
```

```python
#А роза упала на лапу Азора 2.0


def is_palindrome(number):
    number = str(number)
    reversed_number = number[::-1]
    return number == reversed_number

# Input
number = int(input("Enter a natural number: "))

# Check if Palindrome
if is_palindrome(number):
    print("YES")
else:
    print("NO")
```

```python
#Чётная чистота


def remove_even_digits(number):
    purified_number = ""
    for digit in str(number):
        if int(digit) % 2 != 0:
            purified_number += digit
    return int(purified_number)

# Input
number = int(input("Enter a natural number: "))

# Remove Even Digits
purified_number = remove_even_digits(number)

# Output
print(f"Purified number: {purified_number}")
```

```python
#Простая задача 3.0

import math

def is_prime(number):
    if number <= 1:
        return False

    # Check divisibility up to the square root of the number
    for i in range(2, int(math.sqrt(number)) + 1):
        if number % i == 0:
            return False

    return True

# Input
N = int(input("Enter the number of values: "))
numbers = []
for _ in range(N):
    number = int(input("Enter a number: "))
    numbers.append(number)

# Count Prime Numbers
count = 0
for number in numbers:
    if is_prime(number):
        count += 1

# Output
print(f"Total prime numbers: {count}")
```
